```python
import sys
import os
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "..")))
import numpy as np
import matplotlib.pyplot as plt
from core.black_scholes import black_scholes_call

S0 = 100
K = 100
```

```python
vols = np.linspace(0.01, 0.6, 30)
prices_vol = [black_scholes_call(S0, K, T=1, r=0.05, sigma=σ) for σ in vols]

plt.figure(figsize=(8,5))
plt.plot(vols, prices_vol, marker='o')
plt.title("Option Price vs Volatility (σ)")
plt.xlabel("Volatility (σ)")
plt.ylabel("Call Option Price")
plt.grid(True)
plt.show()
```

```python
maturities = np.linspace(0.01, 3, 30)
prices_T = [black_scholes_call(S0, K, T=t, r=0.05, sigma=0.2) for t in maturities]

plt.figure(figsize=(8,5))
plt.plot(maturities, prices_T, marker='s', color='green')
plt.title("Option Price vs Time to Maturity (T)")
plt.xlabel("Maturity (T)")
plt.ylabel("Call Option Price")
plt.grid(True)
plt.show()
```

```python
rates = np.linspace(0.0, 0.2, 30)
prices_r = [black_scholes_call(S0, K, T=1, r=ri, sigma=0.2) for ri in rates]

plt.figure(figsize=(8,5))
plt.plot(rates, prices_r, marker='^', color='orange')
plt.title("Option Price vs Risk-Free Rate (r)")
plt.xlabel("Risk-Free Rate (r)")
plt.ylabel("Call Option Price")
plt.grid(True)
plt.show()
```

## Sensitivity Analysis

This notebook visualizes how the Black-Scholes call option price reacts to changes in key parameters:

- **Volatility (σ)**: Option price increases non-linearly as volatility rises.
- **Maturity (T)**: Longer maturities typically increase the value of call options.
- **Risk-Free Rate (r)**: Call option prices grow as r increases, due to higher present value of expected gains.

These relationships confirm classic financial theory and provide intuition for real-world pricing sensitivity.
